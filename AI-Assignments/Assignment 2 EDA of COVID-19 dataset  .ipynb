Univariate Analysis

```python
print(dataframe.describe())
```

```python
print(dataframe['Deaths'].value_counts())
```

```python
sns.histplot(dataframe['Deaths'].dropna(), kde=True)
plt.title('Deaths')
plt.show()
```

```python
sns.countplot(x='Deaths', data=dataframe)
plt.title('Deaths')
plt.show()
```

 Bivariate Analysis

```python
sns.scatterplot(x='Deaths', y='Recovered', data=dataframe)
plt.title('Deaths vs Recovered')
plt.show()
```

```python
sns.boxplot(x='Deaths', y='Recovered', data=dataframe)
plt.title('Deaths by Recovered')
plt.show()
```

Multivariate Analysis

```python
sns.pairplot(dataframe[['Deaths', 'Recovered', 'Active', 'New cases']].dropna(), hue='Deaths')
plt.show()
```

```python
sns.scatterplot(x='Deaths', y='Recovered', hue='Deaths', style='Active', data=dataframe)
plt.title('Deaths vs Recovered by Recovered and Active')
plt.show()
```
